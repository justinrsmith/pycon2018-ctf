```python
from pandas import read_csv, merge
```

```python
df_prices = read_csv('join_product_prices.csv')
```

```python
df_discounts = read_csv('join_product_discounts.csv')
```

```python
df_prices['Original Price'] = (df_prices['Original Price'].replace( '[\$,)]','', regex=True )
                   .replace( '[(]','-',   regex=True ).astype(float))
df_prices.head()
```

```python
df_discounts['Discounted Price'] = (df_discounts['Discounted Price'].replace( '[\$,)]','', regex=True )
                   .replace( '[(]','-',   regex=True ).astype(float))
df_prices.head()
```

```python
df_merge = merge(df_prices, df_discounts, how='left', left_on=['Product ID'], right_on=['Discounted Product ID'])
```
